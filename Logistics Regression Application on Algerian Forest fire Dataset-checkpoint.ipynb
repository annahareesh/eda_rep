### Checking Outliers

```python
plt.figure(figsize =(15,15))
plt.suptitle('Uni Variate Analysis of Numerical Features', fontsize = 20 )

for i in range(0, len(feature)):
    plt.subplot(5,3,i+1)
    sns.boxplot(x= df[feature[i]])
    plt.xlabel(feature[i])
    plt.tight_layout()
```

Here we can find many Outlier values in the dataset

```python
col = ['Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI',
       'FWI']
```

```python
def find_boundaries(df, variable, distance):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)
    lower_boundary = df[variable].quantile(0.25) - (IQR*distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR*distance)
    return upper_boundary, lower_boundary
```

```python
for i in col:
    upper_boundary, lower_boundary = find_boundaries(df, i, 1.5)
    outliers = np.where(df[i] > upper_boundary, True, np.where(df[i] < lower_boundary, True, False))
    df[i] =  df.loc[~outliers, i]        
```

```python
plt.figure(figsize =(15,15))
plt.suptitle('Uni Variate Analysis of Numerical Features', fontsize = 20 )

for i in range(0, len(feature)):
    plt.subplot(5,3,i+1)
    sns.boxplot(x= df[feature[i]])
    plt.xlabel(feature[i])
    plt.tight_layout()
```

```python
# importing sklearn

from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LogisticRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor 
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
X_scaled
```

```python
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(X_scaled,i) for i in range(X_scaled.shape[1])]
vif["Features"] = X.columns

#let's check the values
vif
```

1. VIF equal to 1 = variables are not correlated
2. VIF between 1 and 5 = variables are moderately correlated
3. VIF greater than 5 = variables are highly correlated

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y, test_size= 0.25, random_state = 355)
```

```python
log_reg = LogisticRegression()

log_reg.fit(x_train,y_train)
```

```python
import pickle
# Writing different model files to file

with open( 'modelForPrediction.sav', 'wb') as f:
    
    pickle.dump(log_reg,f)
    
with open('sandardScalar.sav', 'wb') as f:
    
    pickle.dump(scalar,f)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
# Confusion Matrix
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_negative = conf_mat[1][1]
```

```python
# Breaking down the formula for Accuracy
Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)
Accuracy
```

```python
# Precison
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
# Recall
Recall = true_positive/(true_positive+false_negative)
Recall
```

```python
# F1 Score
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
F1_Score
```

```python
# Area Under Curve
auc = roc_auc_score(y_test, y_pred)
auc
```
